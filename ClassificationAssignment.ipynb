```python
import pandas as pd
```

```python
data=pd.read_csv("CKD.csv")
```

```python
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
independent=data[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dependent=data[['classification_yes']]
```

```python
data["classification_yes"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,train_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

grid_params={"solver":['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky'] ,
             "multi_class":['auto', 'ovr', 'multinomial']}

grid=GridSearchCV(LogisticRegression(),grid_params,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)

ypred=grid.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ypred)
cm
from sklearn.metrics import classification_report
class_report=classification_report(y_test,ypred)
print(class_report)
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

grid_params={"n_estimators":[10,50,100,200],"criterion": ['gini', "entropy", 'log_loss'] ,
             "max_features":['sqrt', 'log2']}

grid=GridSearchCV(RandomForestClassifier(),grid_params,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)

ypred=grid.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ypred)
cm
from sklearn.metrics import classification_report
class_report=classification_report(y_test,ypred)
print(class_report)
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

grid_params={"splitter":['best','random'],"criterion" :['gini', "entropy", 'log_loss'] ,
             "max_features":['auto',  'sqrt', 'log2']}

grid=GridSearchCV(DecisionTreeClassifier(),grid_params,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)

ypred=grid.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ypred)
cm
from sklearn.metrics import classification_report
class_report=classification_report(y_test,ypred)
print(class_report)
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
